```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'/content/population.csv')

# Extract ages from the dataset
Year = data['Year']

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(Year, bins=10, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Distribution of Population over Years')

# Show plot
plt.tight_layout()
plt.show()
```
